```python
import pandas as pd
huge_content = pd.read_csv("plant_based_food_content.csv")
nutri_matter = pd.read_csv("human_daily_nutrition.csv")
code_conversion = pd.read_csv("final_database.csv")
```

```python
nutri_matter.nutrient.to_list()
code_conversion.db_name.to_list()
print(code_conversion.shape)
code_conversion.dropna(subset=['my_name'], inplace=True)
print(code_conversion.shape)
```

```python
len(code_conversion.db_name.to_list())
```

```python
# remove NA rows in orig_source_id and orig_source_name
huge_content.dropna(subset=['orig_food_id','orig_source_id', 'orig_source_name'], inplace=True)
huge_content.orig_food_common_name.value_counts()
huge_content = huge_content[huge_content.orig_source_id.isin(code_conversion.db_name.to_list())]

pivot_df = huge_content.pivot_table(index='orig_food_common_name', columns='orig_source_id', values='orig_max', fill_value=0).reset_index()

# Renaming the index column back to 'food'
pivot_df.rename_axis(None, axis=1, inplace=True)
pivot_df.rename(columns={'orig_food_common_name': 'food'}, inplace=True)

# Display the transformed DataFrame
pivot_df
```
